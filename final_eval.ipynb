### Stability

```python
def plot_stability_bars(df_list, labels=None, title="Diff fuzz session (bug-exception-line)"):
    interesting_counts = []
    unique_crash_counts = []

    for df in df_list:
        df = df.copy()

        # bar 1: total interesting
        interesting_count = df["is_interesting"].sum()
        interesting_counts.append(interesting_count)

        # bar 2: unique crashes = bug_type + exception + line
        has_crash = (
            df["bug_type"].notna() &
            df["exception"].notna() &
            df["line"].notna()
        )

        unique_crashes = (
            df.loc[has_crash, ["bug_type", "exception", "line"]]
            .astype(str)
            .drop_duplicates()
            .shape[0]
        )
        unique_crash_counts.append(unique_crashes)

    if labels is None:
        labels = [f"df_{i}" for i in range(1, len(df_list) + 1)]

    x = np.arange(len(df_list))
    width = 0.35

    plt.figure(figsize=(10, 5))
    plt.bar(x - width/2, interesting_counts, width, label="is_interesting.sum()")
    plt.bar(x + width/2, unique_crash_counts, width, label="Unique Crashes")

    plt.xticks(x, labels)
    plt.xlabel("DataFrames")
    plt.ylabel("Count")
    plt.title(title)
    plt.legend()
    plt.grid(True, axis="y")
    plt.tight_layout()
    plt.show()
```

```python
def count_bug_types(df):
    bug_counts = df['bug_type'].value_counts(dropna=False)
    
    # print("Bug type counts:")
    # print(bug_counts)
    # print("------------------------")
    
    return bug_counts

def count_exceptions(df):
    exception_counts = df['exception'].value_counts(dropna=False)
    
    # print("Exception counts:")
    # print(exception_counts)
    # print("------------------------")
    
    return exception_counts

def count_bug_exception_pairs(df):
    bug_exc_counts = (
        df
        .groupby(['bug_type', 'exception'], dropna=False)
        .size()
        .sort_values(ascending=False)
    )
    
    # print("Bug type + exception counts:")
    # print(bug_exc_counts)
    # print("------------------------")
    
    return bug_exc_counts
```

```python
dfs = [
    (df_json_1, "JSON"),
    (df_ipv4, "IPv4"),
    (df_ipv6, "IPv6"),
    (df_cidrize, "cidrize")
]

fig, axes = plt.subplots(1, 4, figsize=(24, 5))

max_elapsed = 0
prepared = []

for df, title in dfs:
    df = df.copy()

    # Convert timestamp from unix seconds
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")

    # Sort by time
    df = df.sort_values("timestamp").reset_index(drop=True)

    # Keep all 5000 rows, but only define bug signatures where bug info exists
    has_bug = (
        df["bug_type"].notna() &
        df["exception"].notna() &
        df["line"].notna()
    )

    df["bug_signature"] = None
    df.loc[has_bug, "bug_signature"] = (
        df.loc[has_bug, "bug_type"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "exception"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "line"].astype(str).str.strip()
    )

    # Mark only the FIRST occurrence of each unique bug
    df["is_new_bug"] = False
    df.loc[has_bug, "is_new_bug"] = ~df.loc[has_bug, "bug_signature"].duplicated()

    # Cumulative count of unique crashes across all 5000 rows
    df["Unique Crashes"] = df["is_new_bug"].cumsum()

    # Elapsed time in minutes from start of campaign
    df["elapsed_min"] = (
        df["timestamp"] - df["timestamp"].min()
    ).dt.total_seconds() / 60

    max_elapsed = max(max_elapsed, df["elapsed_min"].max())
    prepared.append((df, title))

for i, (df, title) in enumerate(prepared):
    df_new = df[df["is_new_bug"]].copy()

    if len(df_new) == 0:
        axes[i].set_title(f"{title} Crashes")
        axes[i].text(0.5, 0.5, "No unique crashes", ha="center", va="center")
        axes[i].set_xlim(0, max_elapsed)
        axes[i].grid(True)
        continue

    # Plot only the points/line where a new unique bug is discovered
    axes[i].plot(
        df_new["elapsed_min"],
        df_new["Unique Crashes"],
        marker="o",
        linestyle="-",
        label="Unique crashes"
    )

    axes[i].set_title(f"{title} Crashes")
    axes[i].set_xlabel("Minutes since fuzzing start")
    axes[i].set_ylabel("Unique Crashes")
    axes[i].set_xlim(0, max_elapsed)
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
def prepare_unique_crash_curve(dfs, signature_cols, use_time=True):
    """
    Prepare each dataframe by:
    - sorting by timestamp (if use_time=True)
    - computing bug_signature from signature_cols
    - marking first occurrence of each unique bug
    - computing cumulative unique crash count
    - adding x-axis column:
        - elapsed_min if use_time=True
        - execution = row index + 1 if use_time=False

    Parameters
    ----------
    dfs : list of tuples
        Example: [(df_json_1, "JSON"), (df_ipv4, "IPv4")]
    signature_cols : list[str]
        Columns used to define a unique bug signature.
        Example: ["bug_type", "exception", "line"]
    use_time : bool
        True  -> x-axis is elapsed minutes
        False -> x-axis is execution number (row count)

    Returns
    -------
    prepared : list of tuples
        [(prepared_df, title), ...]
    max_x : float or int
        maximum x-axis value across all dfs
    """
    prepared = []
    max_x = 0

    for df, title in dfs:
        df = df.copy()

        if use_time:
            df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
            df = df.sort_values("timestamp").reset_index(drop=True)
        else:
            df = df.reset_index(drop=True)

        # A row is only treated as a valid bug if ALL signature columns are present
        has_bug = df[signature_cols].notna().all(axis=1)

        # Build bug signature only where all required fields exist
        df["bug_signature"] = None
        df.loc[has_bug, "bug_signature"] = (
            df.loc[has_bug, signature_cols]
            .astype(str)
            .apply(lambda col: col.str.strip())
            .agg(" | ".join, axis=1)
        )

        # Mark first occurrence of each unique bug
        df["is_new_bug"] = False
        df.loc[has_bug, "is_new_bug"] = ~df.loc[has_bug, "bug_signature"].duplicated()

        # Cumulative unique crash count over ALL rows
        df["Unique Crashes"] = df["is_new_bug"].cumsum()

        if use_time:
            df["elapsed_min"] = (
                df["timestamp"] - df["timestamp"].min()
            ).dt.total_seconds() / 60
            x_col = "elapsed_min"
        else:
            df["execution"] = df.index + 1
            x_col = "execution"

        max_x = max(max_x, df[x_col].max())
        prepared.append((df, title))

    return prepared, max_x


def prepare_unique_crash_curve_bug_exc_line(dfs, use_time=True):
    """
    Prepare unique crash curve using bug_type + exception + line
    as the bug signature.
    """
    return prepare_unique_crash_curve(
        dfs,
        signature_cols=["bug_type", "exception", "line"],
        use_time=use_time,
    )


def prepare_unique_crash_curve_bug_exc(dfs, use_time=True):
    """
    Prepare unique crash curve using bug_type + exception
    as the bug signature.
    """
    return prepare_unique_crash_curve(
        dfs,
        signature_cols=["bug_type", "exception"],
        use_time=use_time,
    )


def plot_unique_crashes_by_minutes(
    dfs,
    figsize=(24, 5),
    compute_bug_sig=prepare_unique_crash_curve_bug_exc_line
):
    """
    Plot cumulative unique crashes against elapsed minutes.
    Every row is plotted, not only new crash rows.
    """
    prepared, max_x = compute_bug_sig(dfs, use_time=True)

    fig, axes = plt.subplots(1, len(prepared), figsize=figsize)

    if len(prepared) == 1:
        axes = [axes]

    for i, (df, title) in enumerate(prepared):
        axes[i].plot(
            df["elapsed_min"],
            df["Unique Crashes"],
            linestyle="-",
            marker="o",
            label="Unique crashes"
        )
        axes[i].set_title(f"{title} Crashes")
        axes[i].set_xlabel("Minutes since fuzzing start")
        axes[i].set_ylabel("Unique Crashes")
        axes[i].set_xlim(0, max_x)
        axes[i].grid(True)
        axes[i].legend()

    plt.tight_layout()
    plt.show()


def plot_unique_crashes_by_executions(
    dfs,
    figsize=(24, 5),
    compute_bug_sig=prepare_unique_crash_curve_bug_exc_line
):
    """
    Plot cumulative unique crashes against execution count.
    Execution count is just row number (1, 2, 3, ...).
    Every row is plotted, not only new crash rows.
    """
    prepared, max_x = compute_bug_sig(dfs, use_time=False)

    fig, axes = plt.subplots(1, len(prepared), figsize=figsize)

    if len(prepared) == 1:
        axes = [axes]

    for i, (df, title) in enumerate(prepared):
        axes[i].plot(
            df["execution"],
            df["Unique Crashes"],
            linestyle="-",
            marker="o",
            label="Unique crashes"
        )
        axes[i].set_title(f"{title} Crashes")
        axes[i].set_xlabel("Executions")
        axes[i].set_ylabel("Unique Crashes")
        axes[i].set_xlim(1, max_x)
        axes[i].grid(True)
        axes[i].legend()

    plt.tight_layout()
    plt.show()
```

```python
dfs = [
    (df_json_1, "JSON"),
    (df_ipv4, "IPv4"),
    (df_ipv6, "IPv6"),
    (df_cidrize, "cidrize")
]
print("Unique crashes (Dedup with bug_type, exception, line)")
plot_unique_crashes_by_minutes(dfs)
plot_unique_crashes_by_executions(dfs)
```

```python
def plot_unique_exec_against_baseline(
    dfs,
    title="Unique Crashes vs Execution",
    xlim=None,
    signature_cols=["bug_type", "exception", "line"],
    figsize=(3, 3)
):
    prepared, _ = prepare_unique_crash_curve(
        dfs,
        signature_cols=signature_cols,
        use_time=False
    )

    plt.figure(figsize=figsize)

    for df, label in prepared:
        linestyle = ":" if "baseline" in label.lower() else "-"

        plt.step(
            df["execution"],
            df["Unique Crashes"],
            where="post",
            label=label,
            linestyle=linestyle,
            linewidth=2
        )

    plt.title(title)
    plt.xlabel("Execution / Iteration")
    plt.ylabel("Unique Crashes")

    if xlim is not None:
        plt.xlim(1, xlim)

    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_unique_exec_against_baseline_4(
    groups,
    signature_cols=["bug_type", "exception", "line"],
    xlim=None,
    figsize=(18, 5)
):

    fig, axes = plt.subplots(1, 4, figsize=figsize)

    for i, group in enumerate(groups):
        ax = axes[i]

        prepared, _ = prepare_unique_crash_curve(
            group,
            signature_cols=signature_cols,
            use_time=False
        )

        for df, label in prepared:
            linestyle = ":" if "baseline" in label.lower() else "-"

            ax.step(
                df["execution"],
                df["Unique Crashes"],
                where="post",
                label=label,
                linestyle=linestyle,
                linewidth=2
            )

        title = " vs ".join([label for _, label in group])
        ax.set_title(title)
        ax.set_xlabel("Execution")
        ax.set_ylabel("Unique Crashes")
        ax.grid(True)

        if xlim is not None:
            ax.set_xlim(1, xlim)

        ax.legend(loc="upper left")

    plt.tight_layout()
    plt.show()
```

```python
def print_unique_bugs(df, label, signature_cols=["bug_type", "exception", "line"],):
    pd.set_option("display.max_colwidth", None)
    pd.set_option("display.max_columns", None)
    pd.set_option("display.width", None)

    # use helper above
    prepared, _ = prepare_unique_crash_curve(
        [(df, label)],
        signature_cols=signature_cols,
        use_time=True
    )

    df_prepared, label_prepared = prepared[0]

    unique_bugs = df_prepared[df_prepared["is_new_bug"]].copy()

    print(f"\n=== {label_prepared} UNIQUE BUGS ===")
    print(f"Total unique bugs: {len(unique_bugs)}\n")

    cols_to_show = ["timestamp", "bug_signature"] + signature_cols
    cols_to_show = [c for c in cols_to_show if c in unique_bugs.columns]

    display(unique_bugs[cols_to_show])
```
