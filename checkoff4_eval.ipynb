```python
df_json_ledger = pd.read_csv("logs/bug_repro_ledger_5000_json.csv")
print(f'My ledger file has {len(df_json_ledger)} number of rows')
df_ipv4_ledger = pd.read_csv("logs/bug_repro_ledger_5000_ipv4.csv")
print(f'My ledger file has {len(df_ipv4_ledger)} number of rows')
df_cidrize_ledger = pd.read_csv("logs/bug_repro_ledger_5000_cidrize.csv")
print(f'My ledger file has {len(df_cidrize_ledger)} number of rows')
df_ipv6_ledger = pd.read_csv("logs/bug_repro_ledger_5000_ipv6.csv")
print(f'My ledger file has {len(df_ipv6_ledger)} number of rows')
```

```python
dfs = [
    (df_json_ledger, "JSON"),
    (df_ipv4_ledger, "IPv4"),
    (df_cidrize_ledger, "cidrize"),
    (df_ipv6_ledger, "IPv6")
]

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

max_elapsed = 0
prepared = []

for df, title in dfs:
    df = df.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
    df = df.sort_values("timestamp").reset_index(drop=True)
    df["Unique Crashes"] = range(1, len(df) + 1)
    df["elapsed_sec"] = (df["timestamp"] - df["timestamp"].min()).dt.total_seconds()
    df["elapsed_min"] = df["elapsed_sec"] / 60
    max_elapsed = max(max_elapsed, df["elapsed_min"].max())
    prepared.append((df, title))

for i, (df, title) in enumerate(prepared):
    coeffs = np.polyfit(df["elapsed_min"], df["Unique Crashes"], 2)
    trendline = np.polyval(coeffs, df["elapsed_min"])

    axes[i].plot(df["elapsed_min"], df["Unique Crashes"], marker="o", linestyle="-", label="Actual")
    axes[i].plot(df["elapsed_min"], trendline, linestyle="--", label="Trendline")

    axes[i].set_title(f"{title} Crashes")
    axes[i].set_xlabel("Minutes since first unique crash")
    axes[i].set_ylabel("Unique Crashes")
    axes[i].set_xlim(0, max_elapsed)
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
dfs = [
    (df_json, "JSON"),
    (df_ipv4, "IPv4"),
    (df_ipv6, "IPv6"),
    (df_cidrize, "cidrize")
]

fig, axes = plt.subplots(1, 4, figsize=(24, 5))

max_elapsed = 0
prepared = []

for df, title in dfs:
    df = df.copy()

    # Convert timestamp from unix seconds
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")

    # Sort by time
    df = df.sort_values("timestamp").reset_index(drop=True)

    # Keep all 5000 rows, but only define bug signatures where bug info exists
    has_bug = (
        df["bug_type"].notna() &
        df["exception"].notna() &
        df["line"].notna()
    )

    df["bug_signature"] = None
    df.loc[has_bug, "bug_signature"] = (
        df.loc[has_bug, "bug_type"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "exception"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "line"].astype(str).str.strip()
    )

    # Mark only the FIRST occurrence of each unique bug
    df["is_new_bug"] = False
    df.loc[has_bug, "is_new_bug"] = ~df.loc[has_bug, "bug_signature"].duplicated()

    # Cumulative count of unique crashes across all 5000 rows
    df["Unique Crashes"] = df["is_new_bug"].cumsum()

    # Elapsed time in minutes from start of campaign
    df["elapsed_min"] = (
        df["timestamp"] - df["timestamp"].min()
    ).dt.total_seconds() / 60

    max_elapsed = max(max_elapsed, df["elapsed_min"].max())
    prepared.append((df, title))

for i, (df, title) in enumerate(prepared):
    df_new = df[df["is_new_bug"]].copy()

    if len(df_new) == 0:
        axes[i].set_title(f"{title} Crashes")
        axes[i].text(0.5, 0.5, "No unique crashes", ha="center", va="center")
        axes[i].set_xlim(0, max_elapsed)
        axes[i].grid(True)
        continue

    # Plot only the points/line where a new unique bug is discovered
    axes[i].plot(
        df_new["elapsed_min"],
        df_new["Unique Crashes"],
        marker="o",
        linestyle="-",
        label="Unique crashes"
    )

    axes[i].set_title(f"{title} Crashes")
    axes[i].set_xlabel("Minutes since fuzzing start")
    axes[i].set_ylabel("Unique Crashes")
    axes[i].set_xlim(0, max_elapsed)
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
def print_unique_bugs(df, label):
    df = df.copy()

    # Ensure full display
    import pandas as pd
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)

    # Sort properly first
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
    df = df.sort_values("timestamp").reset_index(drop=True)

    # Define bug presence
    has_bug = (
        df["bug_type"].notna() &
        df["exception"].notna() &
        df["line"].notna()
    )

    # Create bug signature
    df["bug_signature"] = None
    df.loc[has_bug, "bug_signature"] = (
        df.loc[has_bug, "bug_type"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "exception"].astype(str).str.strip() + " | " +
        df.loc[has_bug, "line"].astype(str).str.strip()
    )

    # First occurrence logic
    df["is_new_bug"] = False
    df.loc[has_bug, "is_new_bug"] = ~df.loc[has_bug, "bug_signature"].duplicated()

    # Extract unique bugs
    unique_bugs = df[df["is_new_bug"]].copy()

    print(f"\n=== {label} UNIQUE BUGS ===")
    print(f"Total unique bugs: {len(unique_bugs)}\n")

    display(unique_bugs[[
        "timestamp",
        "bug_signature",
        "bug_type",
        "exception",
        "line",
        # "mutated_text", 
        # "mutated_b64" 
    ]])
```
